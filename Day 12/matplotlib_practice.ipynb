# Matplotlib practice

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Plotting a bar graph
x=(1,2)
y=(5,9)
plt.bar(x,y)
plt.show()
```

```python
# Plotting a line graph
x=(5,6,12)
y=(8,7,6)
plt.plot(x,y)
plt.show()
```

```python
# Markers and gridlines
plt.plot(x,y,marker='*') # Marks the points with a symbol(Only predefined symbols)
plt.grid() # Displays grid lines on the graph
plt.show()
```

```python
# Colors, linewidth and linestyles
plt.plot(x,y,linestyle=':') # Displays the graph in dotted line
plt.plot(x,y,linestyle='-.',color='k',linewidth=3) # Plots the line in black coloured '-.' style with width of 3px
plt.show()
```

```python
# Working with datasets
# Histogram
data=pd.read_csv('demo.csv')
# print(data)
plt.hist(data['Age_Group'],edgecolor='red')
plt.show()
```

```python
# Box plot
plt.boxplot(data['Average_Hours'])
plt.show()
```

```python
# Pie-Chart (Used to visualize categorical variables)
count=data['Device'].value_counts()
# plt.pie(count) # Represents the count values in pie-chart
plt.pie(count,labels=count.index,autopct="%1.1f", explode=[0.1,0,0,0]) #"%1.1f - prints the percentage values upto 1 decimal point"
# Explode is used to seperate out a specific section from the pie chart
plt.axis('equal')
```

# Bi-Variate analysis (Working with 2 columns at a time)

```python
data
```

```python
# Scatter Plot
plt.scatter(data['Age_Group'],data['Average_Hours'])
plt.xlabel('Age Group')
plt.ylabel('Average Screen Time')
plt.show()
```

```python
# Line plot
plt.plot(data['Age_Group'],data['Average_Hours'],color='g')
plt.show()
```

```python
# Plotting different types of columns (Numerical and Categorical)
male_screentime=data[data['Gender']=='Male']['Average_Hours']
female_screentime=data[data['Gender']=='Female']['Average_Hours']
```

```python
# Box plot
plt.boxplot([male_screentime,female_screentime],tick_labels=['Male','Female'])
plt.show()
```

```python
screentime_by_device=data.groupby('Device')['Average_Hours'].mean()
screentime_by_device
```

```python
plt.pie(screentime_by_device,labels=screentime_by_device.index,autopct='%1.0f',labeldistance=0.4)
plt.title('Device Average Screentime')
plt.legend()
```

# Multivariate Analysis (2-Numerical,1-Categorical)

```python
#Scatter-plot(Bubble plot)
plt.scatter(data['Age_Group'],data['Average_Hours'],c=data['Device'].map({'Mobile':'violet','Laptop':'blue','Tablet':'green','TV':'orange'}))
# Saving the images to the device
plt.savefig("Bubbleplot.png")
plt.show()
```
